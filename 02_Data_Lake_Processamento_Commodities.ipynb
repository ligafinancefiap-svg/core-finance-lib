Segmento - Commodities - Lê da Bronze e Salva na Silver

Camada Silver

```python
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')

segmento = "commodities"
caminho_bronze = f'/content/drive/MyDrive/DATA_LAKE/01_Bronze/b3_{segmento}_brutos.csv'
caminho_silver = f'/content/drive/MyDrive/DATA_LAKE/02_Silver/b3_{segmento}_limpo.csv'

def processar_silver():
    if os.path.exists(caminho_bronze):
        print(f"⚙️ Iniciando Processamento Silver: {segmento.upper()}")
        df = pd.read_csv(caminho_bronze, index_col=0, parse_dates=True)

        # Limpeza
        df = df[~df.index.duplicated(keep='first')]
        df_limpo = df.ffill()
        df_limpo.columns = [col.replace('.SA', '') for col in df_limpo.columns]

        df_limpo.to_csv(caminho_silver)
        print(f"✅ Dados limpos salvos em Silver.")
    else:
        print("⚠️ Arquivo Bronze não encontrado.")

processar_silver()
```
