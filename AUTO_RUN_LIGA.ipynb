Executa comandos de forma automatizada

```python
from google.colab import drive
import os

# 1. Conexão com o Drive
drive.mount('/content/drive')

# 2. Pasta onde estão os seus notebooks
path = "/content/drive/MyDrive/Colab Notebooks/"

def executar_fluxo(arquivo):
    caminho_total = os.path.join(path, arquivo)
    if os.path.exists(caminho_total):
        print(f"🚀 Rodando: {arquivo}")
        try:
            %run "{caminho_total}"
            print(f"✅ {arquivo} OK")
        except Exception as e:
            print(f"❌ Erro em {arquivo}: {e}")
    else:
        print(f"⚠️ Arquivo não encontrado: {arquivo}")

# 3. LISTA MESTRE - A ordem de execução da sua fábrica
# IMPORTANTE: Verifique se os nomes dos seus arquivos são exatamente estes
fluxo_da_liga = [
    # --- CAMADA 01 (BRONZE) - Extração ---
    "01_Data_Lake_Bancos.ipynb",
    "01_Data_Lake_Commodities.ipynb",
    "01_Data_Lake_FIIs.ipynb",

    # --- CAMADA 02 (SILVER) - Limpeza ---
    "02_Data_Lake_Processamento_Bancos.ipynb",
    "02_Data_Lake_Processamento_Commodities.ipynb",
    "02_Data_Lake_Processamento_FIIs.ipynb",

    # --- CAMADA 03 (GOLD) - Análise Quant ---
    "03_Data_Lake_Analise_Quant_Bancos.ipynb",
    "03_Data_Lake_Analise_Quant_Commodities.ipynb",
    "03_Data_Lake_Analise_Quant_FIIs.ipynb",

    # --- CONSOLIDAÇÃO FINAL ---
    "00_Data_Lake_Consolidação_Portfolio.ipynb"
]

print("🏁 INICIANDO ATUALIZAÇÃO TOTAL DA LIGA FINANCEIRA\n" + "="*50)

for nb in fluxo_da_liga:
    executar_fluxo(nb)

print("\n" + "="*50 + "\n🏆 TUDO PRONTO! Seu patrimônio e análises foram atualizados.")
```
