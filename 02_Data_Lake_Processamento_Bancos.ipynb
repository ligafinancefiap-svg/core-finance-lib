Segmento - Bancos - Lê da Bronze e Salva na Silver

Camada Silver


```python
import pandas as pd
import os
from google.colab import drive

# 1. Conexão com o Google Drive
drive.mount('/content/drive')

# 2. Configuração de Caminhos e Segmento
segmento = "bancos"
caminho_bronze = '/content/drive/MyDrive/DATA_LAKE/01_Bronze/'
caminho_silver = '/content/drive/MyDrive/DATA_LAKE/02_Silver/'

# Criar a pasta Silver caso ela ainda não exista
if not os.path.exists(caminho_silver):
    os.makedirs(caminho_silver)
    print(f"📁 Pasta Silver criada com sucesso.")

def processar_bancos_silver():
    arquivo_origem = f'{caminho_bronze}b3_{segmento}_brutos.csv'
    arquivo_destino = f'{caminho_silver}b3_{segmento}_limpo.csv'

    if os.path.exists(arquivo_origem):
        print(f"⚙️ Iniciando Processamento Silver: {segmento.upper()}")

        # LEITURA (Puxando da Bronze)
        df = pd.read_csv(arquivo_origem, index_col=0, parse_dates=True)

        # LIMPEZA E PADRONIZAÇÃO (Refino Silver)
        # 1. Remove duplicatas de datas
        df = df[~df.index.duplicated(keep='first')]

        # 2. Tratamento de Nulos (Forward Fill)
        # Preenche lacunas de feriados ou falhas de cotação com o último preço disponível
        df_limpo = df.ffill()

        # 3. Limpeza de Tickers (Remove o .SA para um visual profissional)
        df_limpo.columns = [col.replace('.SA', '') for col in df_limpo.columns]

        # SALVAMENTO (Guardando na Silver)
        df_limpo.to_csv(arquivo_destino)
        print(f"✅ Sucesso! Dados limpos salvos em: 02_Silver/b3_{segmento}_limpo.csv")

    else:
        print(f"⚠️ Erro: Arquivo bruto de {segmento} não encontrado na Bronze.")

# Execução do Processamento
processar_bancos_silver()
```
